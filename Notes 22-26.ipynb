# Successive approximations (fixed point iterations)

* solution method for root finding:

$x = F(x) \iff F(x) - x = 0$

* **fixed point equation**

## Main algorithm 

$ x = F(x) \rightarrow x_{n+1} = F(x_{n}) $

* initialise at $x_{0}$
* terminate once successive terms are close enough 

--

* used it with stationary distributions in Markov chains: $F(\psi) = \psi P*




## Will it converge?

* it will from any initial condition if MC is aperiodic and irreducible
* other general equations? 
    - needs to be a contraction mapping 
    
Example:
$ \frac{1}{2} - exp(-(x-2)^{2}) = 0$

$ F(x) = x - exp(-(x-2)^{2}) + \frac{1}{2}$ 

because $x = F(x)$. Then:

$x_{i+1} = x_{i} - exp(-(x_{i}-2)^{2}) + \frac{1}{2}$

```python
eq = lambda x: - np.exp(-(x - 2)**2) + .5
F = lambda x: x - np.exp(-(x - 2)**2) + .5
dF = lambda x: 1 + np.exp(-(x - 2)**2) * 2 * (x - 2) # derivative 
```

```python
import numpy as np
```

```python
def solve_sa(F, x0, tol = 1e-6, maxiter = 100, callback = None):
    
    for i in range(maxiter):
        x1 = F(x0)
        err = np.amax(np.abs(x0 - x1)) # allows x to be an array
        if callback != None: callback(iter = i, err = err, x = x1, x0 = x0)
        if err < tol:
            break
        x0 = x1
    else:
        raise RuntimeError('Failed to converge in %d iterations' % maxiter)
    return x1
```

```python
def printme(**kwargs):
    print('iter %d, x = %1.10f (err = %1.3e)'%(kwargs['iter'],kwargs['x'],kwargs['err']))
x = solve_sa(F, x0 = 1.0, tol = 1e-10, callback = printme)
print('Residual in the original equation = %1.5e' % eq(x))
# try different starting points, including >2.8
```

* error decreases in a linear fashion - not a fast equation
* if we change the starting value, then there are a lot more iterations needed - we need a good starting point, which is difficult to find

## Multivariate example: platform market equilibrium 

* multivariate fixed point 

--

* platofrm is a product/ serice with network effects (value is generated from interaction of users)
    - depends on the size of the user base
    - i.e. game consoles, social media, etc. 
* because of the network effects, platform markets tend to monopolization (tipping markets)
    - opposite forces could exist: heterogeneous preferences (of users) or congestion 
    - Vimeo targets particular class of creaters, rather than YouTube 

### The model

* $m$ products 
* unit mass of consumers of $n$ types with different preferences over the product 
* $p_{j}$ are the fractions of consumer types in the population 
* utility: $u_{ij} = c_{ij} + s_{i}$
    - $c_{ij}$ are valuations of each production by each consumer type
    - $s_{i}$ are market shares of each product; greater the market share the better the experience for users 
    
--

* standard random utility framework with logit choice probabilities 
    - value of choice $U_{ij} = u_{ij} + \epsilon[i]$ is affected by the random factors 
        * total utility has deterministic component (from above) and a random component (associated with the platform; special deals, etc. that the users knows but not the econometrician)
    - elements of $\epsilon$ are extreme value type 1 (EV1) random components of the utility, iid between consumers and products:
   
$P_{ij} = \frac{exp(u_{ij})}{\sum_{k=1}^{m}exp(u_{kj})} = \frac{exp(u_{ij} - \alpha)}{\sum_{k=1}^{m}exp(u_{kj} - \alpha)}$

* logit prob. of users type $j$ of choosing product tpye $i$
    - divide each side of fraction by $\epsilon^{\alpha}$; to ensure numerical stability 
    
--

Market shares 

* assume choice is made once
* under unit mass assumptions:
$s_{i} = \sum_{j=1}^{n}p_{j}P_{ij} = P \cdot p$ (simple matrix product)

-- 

Fixed point equation

$u_{ij} = c_{ij} + \sum_{t=1}^{n}p_{t}\frac{exp(u_{it})}{\sum_{k=1}^{m}exp(u_{kt})}$

* utility is calculated from all other utilities
    - denominator: summing over all of the choices that the user could take, conditional on type $t$
    - numerator: type $i$ in the numerator 
* we can write $u_{ij}$ for each $i$ and each $j$:
    - $mn$ fixed point equations 
    - fixed point in the ($mn$)-dimensional space of $u_{ij}$

```python
class model:
    
    def __init__(self, m = 2, n = 2):
        self.m, self.n = m, n
        self.c = np.random.uniform(size = (m, n)) # not working with data, so just use random uniform for c 
        self.p = np.random.dirichlet(np.ones(n)) # population composition (produces fractions/ choice probabilities sum to1)
        
    def __repr__(self): # object representation
        return 'Number of platform products = {:d}\nNumber of consumer types = {:d}\nPopulation composition = {}\nUtilities:\n{}'.format(self.m,self.n,self.p,self.c)        
        
        
    def ccp(self, u):
        '''Conditional choice probabilities.
        
        Conditional on attributes of choices and the consumer types.
        u is the vector of utilities (we need to converge on this); need to treat as a vector
        
        
        
        
        '''
        u = np.asarray(u).reshape((self.m, self.n)) # reshape into correctly shaped matrix (choices in rows, types in columns)
        u = u - np.amax(u, axis = 0) # de-max by column (maximum values among choices for each consumer type) [for numerical stability]
        e = np.exp(u) 
        esum = e.sum(axis = 0) 
        return e / esum # matrix of choice probabilities P (using the formula)
    
    def shares(self, pr):
        out = pr @ self.p # one-dim vector: probabilities multiplied by population composition
        return out[:,np.newaxis] # converting 1D vector into a column vector (adding new dimension)
    
    def F(self, u):
        pr = self.ccp(u)
        sh = self.shares(pr)
        u1 = self.c + sh # updated utility
        
        return u1.reshape(self.m * self.n) # want a 1d array (rather than m*n)
```

```python
def printiter(**kwargs):
    print('iter %d, err = %1.3e'%(kwargs['iter'],kwargs['err']))
md = model(m=3,n=2)
print(md)
print('SA from zero utilities:')
x = solve_sa(md.F,x0=np.zeros(md.n*md.m),callback=printiter)
print('Equilibrium found!')
ccp = md.ccp(x)
shares = md.shares(ccp).squeeze()  # make one-dim array
print('Equilibrium choice probabilities:',ccp,'Equilibrium market shares:',shares,sep='\n')
```

* population composition: 82% are of type 1
* consumers of type 1 (column 1) prefers third product, and type 2 prefers second product (predict movement into these two products 
* eqm choice probabilities: type 1 chooses them almost equally, etc.

# Multivariate Newton-Raphson method 

* iterative solver 
* $x_{i+1} = x_{i} - \frac{f(x_{i})}{f'(x_{i})}$
* fast, quadratic convergence 

Newton step:
* if write Taylor series expansion around the solution
* take first two terms, assuming f(x) is the solution, and let $x_{0} = x_{i}$ and $x = x_{i+1}$

$0 = f(x) = f(x_{i}) + f'(x_{i})(x_{i+1} - x_{i})$ (root-finding algorithm)

after solving for $x_{i+1}$

```python
def newton(fun,grad,x0,tol=1e-6,maxiter=100,callback=None):
    '''Newton method for solving equation f(x)=0
    with given tolerance and number of iterations.
    Callback function is invoked at each iteration if given.
    '''
    for i in range(maxiter):
        x1 = x0 - fun(x0)/grad(x0) # ARGUMENTS ARE THE FUNCTION AND THE DERIVATIVE 
        err = abs(x1-x0)
        if callback != None: callback(iter=i,err=err,x0=x0,x1=x1,fun=fun)
        if err<tol: break
        x0 = x1
    else:
        raise RuntimeError('Failed to converge in %d iterations'%maxiter)
    return x1
```

```python
f = lambda x: -4*x**3+5*x+1
g = lambda x: -12*x**2+5
x = newton(f,g,x0=-2.5,maxiter=7)
print('Solution is x=%1.3f, f(x)=%1.12f' % (x,f(x)))
```

```python
def newton_pic(f,g,x0,    # function, gradient, initial point
               a=0,b=1,   # bounds for the picture
               **kwargs):  # additional parameters passed to solver
    '''Illustrate the Newton method iterations'''
    xd = np.linspace(a,b,1000)       # x grid
    plt.plot(xd,f(xd),c='red')       # plot the function
    plt.plot([a,b],[0,0],c='black')  # plot zero line
    ylim=[f(a),min(f(b),0)]
    def plot_step(**kwargs):
        plot_step.counter += 1
        x0,x1 = kwargs['x0'],kwargs['x1']
        f0 = kwargs['fun'](x0)
        plt.plot([x0,x0],[0,f0],c='green')
        plt.plot([x0,x1],[f0,0],c='green')
    plot_step.counter = 0
    try:
        newton(f,g,x0,callback=plot_step,**kwargs)
        print('Converged in %d steps'%plot_step.counter)
    except RuntimeError:  # catch the Runtime convergence error
        print('Failed to converge in %d iterations!'%plot_step.counter)
    plt.xlim((a,b))
    plt.show()
```

## Algorithm 

$f(x) \rightarrow max$

1. Take starting value x0, define a region of search: $I = (x0 - a, x0 + b)$
2. Impose on $I$ a discrete grid consisting of point $x_{i}$
3. Compute $f(x_{i})$ for all $i$
4. Return the maximum as a result

# Newton-Raphson with bounds 

* models of discrete choice in IO:
    - mixed strategy equilibrium in a two player game with binary actions 
    - calculates the BR of one player 
    
$f(x) = a \log(x) + b \log(1 - x) + c = 0$, $ab < 0$

* $x \in (0, 1)$ are strict limits; any algorithm would break if it steps outside
* we know there is one solution for any $a, b, c$
    - assume $a > 0$
* $f$ is continuous in domain
* $f' \geq 0$ everywhere in domain; monotonically increasing 
* when $x \rightarrow 0$ from the right, then $f(x) \rightarrow \infty$, etc.

$ \Rightarrow $ $f$ only crosses the x-axis once - the solution should be easy to find... however, the bounds are restrictive, and the function may because difficult to implement in the NR method because of the logs. 

1. Newton can overshoot domain easily 
2. Bisections may take forever to converge 
    - solutions may be close to the bounds 
3. Successive approximations 

## Newton-bisection hybrid algorithm 

1. Initialize with interval ($a, b$) such that the function has opposite signs at the ends, and starting value x0
2. Compute the Newton stop and check if it remains within the bounds 
3. If yes, make Newton step and continue to the next iteration
4. If no, make a bisection step and continue to next iteration
5. Iterate until convergence, or maximum number of iterations is reached without convergence 

```python
np.seterr(all=None, divide='ignore', over=None, under=None, invalid='ignore')  # turn off log(0) warning
```

```python
a, b, c = 10, -1, 500 # simplest values 
f = lambda x: a * np.log(x) + b * np.log(1 - x) + c
g = lambda x: a / x - b / (1 - x)
```

```python
xd = np.linspace(0, 1, 100)
plt.plot(xd, f(xd), c = 'r')
plt.plot([0,1], [0,0], c = 'grey')
plt.show()
```

```python
def newton_bounded(fun, grad, x0 = None, a = 0, b = 1, tol = 1e-6, maxiter = 100, callback = None):
    '''Implementation of Newton algorithm with bounds'''
    
    sa, sb = np.sign(fun(a)), np.sign(fun(b)) # check if the signs are opposite
    if sa * sb > 0:
        raise ValueError('Function has the same sign at the bounds')
        
    x0 = (a + b) / 2 if x0 == None else x0
        
    for i in range(maxiter):
        f0 = fun(x0)
        
        newt = x0 - f0/grad(x0) # Newton step 
        
        if not a <= newt <= b: 
            # need bisection step 
            a, b = (x0, b) if np.sign(f0) == sa else (a,x0)
            x1 = (a + b) / 2
            step = 'bisection'
        else: 
            # do not need bisection step
            x1 = newt
            step = 'newton'
            
        err = abs(x1-x0)
        
        if callback != None: callback(iter = i, type = step, err = err, x0 = x0, x1 = x1, a = a, b = b)
            
        if err<tol: 
            break
        x0 = x1
    else:
        raise RuntimeError('Failed to converge in %d iterations' % maxiter)
        
    return x1  
```

## Modelling market equilibrium 

* demand and supply (Marshallian cross)

The supply curve:
* once you get to the threshold, the price needs to rise even further to justify supplying more 

```python
def supply(p):
    '''Vectorized function to compute supply'''
    const = 5 * np.arange(10) # cost thresholds 
    p = np.asarray(p).reshape((1, -1)) # row vector 
    s = np.kron(np.ones((const.size, 1)), p) # repeat rows 
    for i, c in enumerate(const):
        s[i,:] = np.log(s[i,:] * .2 - c) + c + 3
    return np.nanmax(s, axis = 0)

def demand(p):
    '''Vectorized function to compute demand'''
    res = np.asarray(100/p + 1/(1+np.exp(p-75))
                           + 2/(1+np.exp(p-50))
                           + 3/(1+np.exp(p-25)))
    res[res>20] = np.nan  # avoid large numbers for plots
    return res

def plotline(fun,lbl=None,ax=None,**kwargs):
    # quick plotting function on given axis
    if ax==None:
        ax = plt.axes()
    pr = np.linspace(0,100,1000)
    ax.plot(fun(pr),pr,**kwargs)
    if lbl:
        ax.text(fun(60)+1,60,lbl,fontsize=12)
    ax.set_xlabel('Quantity')
    ax.set_ylabel('Price')
    return ax

ax = plotline(demand,lbl='Demand',c='b')
plotline(supply,lbl='Supply',c='r',ax=ax)
plt.show()
```

### Computing market equilibrium: straightforward approach 

* $D = S$, and then use the solver (bisection, etc.)
    - bisection will calculate the intersection (function) many times, so slow
    - Newton: need to calculate the derivative for the function $D - S = 0$, so hard (kinks too)

### Precomputation approach 

1. Approximate supply and demand curves with piecewise linear curves 
    - fixed and known number of calculations of supply and demand 
    - computational load vs. accuracy 
2. Compute the intersection of the piecewise linear approximations 

### Polyline class

* holds the coordinates of the computed points of the line
    - grid of $x$ and the $f(x)$
* implements simple operations on the lines
    - graphing too
* contains methods to compute intersection points 

--

* __init__: three arguments: $x$, $y$ [$y = f(x)$], and optional label 
* __repr__: string representation
* __call__: return interpolated values of $f(x)$ for the passed array $x$
* __add__: concatenate polylines
* **plot**
* **intersect**
* **self_intersect**: routine to find intersections within itseld (i.e. if the function loops)

```python
class polyline():
    
    def __init__(self, x, y, label = 'no label'):
        '''Initializer of the polyline object'''
        
        self.x, self.y = np.asarray(x, dtype = float).ravel(), np.asarray(y, dtype = float).ravel()
        # want self to be the object of attribute
        # ravel makes any size array into a 1d array (like reshaping and then squeezing)
        
        if self.x.size != self.y.size:     # size gives number of elements 
            raise ValueError('The input arguments to polyline should be the same length')
        self.label = '{:s}'.format(label) # to convert to string label
        
    def __len__(self):
        '''Length of the polyline'''
        return self.x.size # know x and y are the same size 
        
        
    def __repr__(self):
        '''String representation of polyline'''
        return 'Polyline with {} points labelled "{}"\nx = {}\ny={}'.format(len(self), self.label, self.x, self.y)
        
        return 'Polyline with {} points labeled {}'.format()
    
    
    def __call__(self,x):
        '''Returns the values of the polyline computed for given points x'''
        
        interp = interpolate.interp1d(self.x, self.y, bounds_error = False, fill_value = np.nan) # is a fun
        return interp(x)
    
    
    def __add__(self, other):
        '''Overload of + to concatenate two polylines'''
        if not isinstance(other, polyline):
            raise TypeError('Only polyline can be added to another polyline')
        x = np.concatenate((self.x, other.x))
        y = np.concatenate((self.y, other.y))
        label = self.label + ' + ' + other.label
        return polyline(x, y, label = label)
    
    
    def plot(self, ax = None, **kwargs):
        '''Makes a plot of polyline'''
        if not ax:
            ax = plt.axes()
        ax.plot(self.x, self.y, **kwargs)
        return ax 
    
    def intersect(self, other):
        '''Returns the 2D numpy array of intersection points (two coordinates) in a row'''
        if not isinstance(other, polyline):
            raise TypeError('Only polyline can be added to another polyline')  
        ub = (len(self) - 1) * (len(other) - 1)
        out = np.empty((ub, 2))  # output; space for the intersection points 
        k = 0 # number of intersection points
        for x11, y11, x12, y12  in zip(self.x, self.y, self.x[1:], self.y[1:]):   # loop over segments, which is a collection of four points (coordinates)
            # zipping creates the pairs 
            # [1:] starts us at the next point 
            # zip exhausts when the shortest list is done 
            for x21, y21, x22, y22  in zip(other.x, other.y, other.x[1:], other.y[1:]):
                pp = self.__intersect_segments(x11, y11, x12, y12, x21, y21, x22, y22)
                if pp:
                    out[k, 0], out[k, 1] = pp
                    k += 1
                    
        out = out[:k,:]
    
        return out
        
        
    def self_intersect(self):
        '''Return the 2-dim numpy array of self-intersection points (in row)'''
        ub = (len(self)-1)**2
        out = np.empty((ub,2))  # space for the intersection point
        k = 0  # number of intersection points so far
        for x11,y11,x12,y12 in zip(self.x,self.y,self.x[1:],self.y[1:]):
            for x21,y21,x22,y22 in zip(self.x,self.y,self.x[1:],self.y[1:]):
                pp = self.__intersect_segments(x11,y11,x12,y12,x21,y21,x22,y22)
                if pp:
                    out[k,0],out[k,1] = pp
                    k += 1
        out = out[:k,:]
        return out   
        
        
            
    def __intersect_segments(self,*args): # __ means function is only accessible from class 
        '''Find intersection point of a segment, or return None'''
        # unpack parameters
        x11,y11,x12,y12=args[:4] #first line segment
        x21,y21,x22,y22=args[4:] #second line segment
                # check if segmets are identical
        if np.all(np.abs(np.asarray(args[:4])-np.asarray(args[4:]))<1e-10):
            return None
        # check if segmets share a point
        if np.all(np.abs(np.asarray(args[:2])-np.asarray(args[4:6]))<1e-10):
            return None
        if np.all(np.abs(np.asarray(args[:2])-np.asarray(args[6:]))<1e-10):
            return None
        if np.all(np.abs(np.asarray(args[2:4])-np.asarray(args[4:6]))<1e-10):
            return None
        if np.all(np.abs(np.asarray(args[2:4])-np.asarray(args[6:]))<1e-10):
            return None
                # bounding box check: whether intersection is possible in principle
        bb = ( min(x11,x12)<max(x21,x22) and min(x21,x22)<max(x11,x12) and
               min(y11,y12)<max(y21,y22) and min(y21,y22)<max(y11,y12) )
        if not bb:
            return None
        # form system of equations
        A = np.array([[x12-x11,0,-1,0],[0,x22-x21,-1,0],[y12-y11,0,0,-1],[0,y22-y21,0,-1]])
        b = np.array([-x11,-x21,-y11,-y21])
        t1,t2,x0,y0 = np.linalg.solve(A,b)
        if 0 <= t1 <= 1 and 0 <= t2 <= 1:
            return x0,y0
        else:
            return None
```

```python
# write unit tests here
import unittest
# see examples at https://docs.python.org/3/library/unittest.html

class TestPolyline(unittest.TestCase):
    """Tests for the polyline class"""

    
    def test_init(self):
        '''Test for the initializer'''
#         polyline([1, 2,],[3, 4,])  # check different data passed as x and y
        self.assertEqual(str(polyline([1, 2, 5, 6],[[5, 6, 5, 6]])),
                        'Polyline with 4 points labelled "no label"\nx = [1. 2. 5. 6.]\ny=[5. 6. 5. 6.]')
        
        self.assertEqual(str(polyline([[1], [2], [5], [6]],[[5, 6, 5, 6]])),
                        'Polyline with 4 points labelled "no label"\nx = [1. 2. 5. 6.]\ny=[5. 6. 5. 6.]')
        
        self.assertEqual(str(polyline(np.array([[1], [2], [5], [6]]),[[5, 6, 5, 6]])),
                        'Polyline with 4 points labelled "no label"\nx = [1. 2. 5. 6.]\ny=[5. 6. 5. 6.]')
        self.assertRaises(ValueError, polyline, [1], [2, 4])
        self.assertRaises(ValueError, polyline, [1, 2,], 'string')
        self.assertRaises(ValueError, polyline, [1], [2], 6758)
        
        
    def test_call(self):
        a = polyline(np.array([[1], [2], [5], [6]]),[[5, 6, 5, 6]])
 # data 0 is interpolated data 
        dt0 = a([2.3,4,5,23])
        dt1 = np.asarray([5.9,5.33333333,5.,np.nan])
        self.assertAlmostEqual(dt0[:3].sum(),dt1[:3].sum())
        self.assertTrue(np.isnan(dt0[3]))
        
        
    def test_add(self):
        a = polyline([0, 0, 1, 2], [9, 9, 3, 4])
        b = polyline([9, 9, 1, 2], [9, 8, 7, 8])
        c = a + b         
        with self.assertRaises(TypeError):
            a + 'string'
            a + 45
        self.assertListEqual(list(c.x), [0, 0, 1, 2, 9, 9, 1, 2])
        self.assertListEqual(list(c.y), [9, 9, 3, 4, 9, 8, 7, 8])
        
        
    def test_intersect(self):
        a = polyline([0, 0, 1, 2], [9, 9, 3, 4])
        b = polyline([9, 9, 1, 2], [9, 8, 7, 8])       
        with self.assertRaises(TypeError):
            a.intersect('string')
            a.intersect(45)

# this is the way to run tests
if __name__ == '__main__':
    # tweaking for Jupyter Notebook
    unittest.main(argv=['first-arg-is-ignored'], exit=False)
    # from command line
    # unittest.main()
```
