### Tiling with dominoes example

* $3 \times n$ board: find the number of ways to fill it with $2 \times 1$ dominoes
* break into smaller sub-problems:
    - at any possible stage, we can have:
    
    1. Completely filled: $A_{n}$ (number of ways to tile this shape)
    2. Corner at top right empty: $B_{n}$
    3. Corner at top bottom empty: $C_{n}$ (can see that other configurations are not possible)
    
--

* to get to $A_{n}$, it is a function of the other shapes
    - $A_{n} = A_{n-2} + B_{n-1} + C_{n-1}$ (and then add the last remaining *single* dominoes to complete whole rectangle
    - $B_{n} = A_{n-1} + B_{n-2}$ (i.e. two ways to get to the shape $B_{n}$
    - $C_{n} = B_{n}$ (just as above)
* now we have recursion:

$A_{n} = A_{n-2} + 2B_{n-1}$

$B_{n} = A_{n-1} + B_{n-2}$

```python
def WaysTileDominoes(n):
    A, B = [0] * (n + 1), [0] * (n + 1)
    A[0] = 1 # one way to tile a 3x0
    A[1] = 0 # no way to tile a 3x1
    B[0] = 0 # no way to tile a 3x0 without a corner
    B[1] = 1 # one way to tile a 3x1 without a corner (just one tile and a bit left over)
    for i in range(2, n + 1): # loop over 2,3,...,n
        A[i] = A[i - 2] + 2 * B[i - 1]
        B[i] = A[i - 1] + B[i - 2]
    return A[n]
```

```python
for n in range(1, 10):
    print('There are', WaysTileDominoes(n),'ways to tile the 3 by', n, 'board')
```

### Deal or no deal example

* player presented with $n$ boxes with prizes inside
* prizes $x_{1},...,x_{n}$ are known but it is not known where
* choose box to be removed from game
* otherwise, can walk away from game and get random prize chosen from remaining boxes 

What is the optimal strategy to maximise expected reward?

* expected reward is: $\frac{x + y + x}{3}$ (stopping the game)
* expected reward is: $\frac{1}{3}\cdot\frac{x + y}{2} + \frac{1}{3}\cdot\frac{x + z}{2} + \frac{1}{3}\cdot\frac{y + z}{2}$ (from continuation of game)
* implies that at each stage, the player will be indifferent between stopping or continuing the game
$\Rightarrow$ optimal strategy is to stop at the first round and take random box

Assume loss aversion utility:

$U$(reward, reference) = reward - $\theta \cdot \mathbb{1}\{reward < reference\}$

* if award is less than the reference level, there is a cut in utility 
* assume it is updated endogeneously; at each round it is equal to the foregone option of stopping the game in the previous round 

--

1. Maximum number of rounds?
    - given by the finite number of boxes
2. What does the optimal choice depend on in each game?
3. What is complete strategy of game? 
    - strategy: like a rulebook of optimal actions in each possible state of the game 
    
--

Let $B$ denote the set of remaining $n$ boxes. Denote $r$ the reference level. The maximum expected reward (choice of stopping and continuing):

$V(B, r) =$ max$\Big[\frac{1}{n}\sum_{b \in B}U(b, r); \sum_{b \in B}\frac{1}{n}V\big(B \backslash b, \frac{1}{n}\sum_{d \in B}U(d, r)\big)\Big]$

$V(B, r) =$ max$\Big[R_{r}; \sum_{b \in B}\frac{1}{n}V(B \backslash b, R_{r})\Big]$

* if we stop the game, then we get the average of the remaining boxes 
* if continue, we have the set of boxes remaining after we throw one ($b$) away, and the reference is what we could have got if we stopped ($R_{r}$)
* we can see the recursion

```python
def expected_reward(boxes,ref=None,loss_aversion_param=.1,verbose=True):
    '''Compute the expected reward from the game with given boxed prizes and the reference level'''
    n = len(boxes)  # number of boxed remaining
    ref = sum(boxes)/n if ref is None else ref  # default reference level
    # reward if stopping
    current = [b - loss_aversion_param*(b<ref) for b in boxes] # utilities of the prizes for each 
    current = sum(current)/n  # average (current reward to compare to continuation reward)
    # expected reward if continuing the game
    if n==1:
        next = current  # the same if only one box left
    else:
        next = 0.  # initialize
        for i in range(n):  # accumulate expected reward component by component
            next += expected_reward(boxes[:i]+boxes[i+1:],current,loss_aversion_param,verbose) # the list without the box i
        next /= n  # take average
    if verbose:
        print('%-20s'%''.join('[{}]'.format(b) for b in boxes),end='')  # this round boxes
        print(' ref={:<6.3f}'.format(ref),end='')                       # this round reference level
        if n>1:
            print(' reward if stop={:<6.3f} if continue={:<6.3f}'.format(current,next),end='')  # rewards for two decisions
            print('  >> {}!'.format('stop' if current >= next else 'continue'))                 # best decision
        else:
            print(' reward={:<6.3f}'.format(current))   # reward in case of last box left
    return max(current,next)  # reward from the optimal choice
```

## Bellman Principle of Optimality 

* given initial state/ conditions, an optimal strategy must be that the decisions after the initial decisions is also optimal 
* breaking into sequence of sub-problems 
    - initial and future decision problem
    - can be computed through **backward induction** (when horizon is finite)

### Bellman equation

$V(\textrm{state}) = \max_{\textrm{decisions}}\Big[U(\textrm{state}, \textrm{decision}) + \beta \mathbb{E} \big\{V(\textrm{next state}) \big| \textrm{state}, \textrm{decision}\big\}\Big]$

* the value function (maximum attainable expected reward) is a function of the per-period/ instantaneous reward and the next-period value function
* in deterministic case, can remove $\mathbb{E}$


For finite horizon problems, in the final period $t = T$, we only maximise the current payoff because:
$V(T + 1) = \mathbb{0}$

### Inventory dynamics problem

* discrete time
* inventory $x_{t} \geq 0$ measured in discrete units 
* $d_{t}$ is stochastic demand
* $q_{t}$ is order of new inventory 
* $p$ is profit per unit of supplied good
* $c$ is fixed cost of ordering any amount
* $r$ is the cost of storing one unit of good

Sales are given by $s_{t} = \min\{x_{t}, d_{t}\}$ (either you meet the demand, or all the stock you have). Inventory to be stored: $k_{t} = \max\{x_{t} - d_{t}, 0\} + q_{t} = x_{t+1}$ (remaining units after sales, plus new orders. Profit:

$\pi_{t} = p \cdot s_{t} - r \cdot x_{t+1} - c \cdot (\textrm{orders made})$

$\pi_{t} = s_{t} p - k_{t} r - c\mathbb{1}\{q_{t} > 0\}$

Assume only positive amounts can be ordered, assume $\sigma$ is set of all feasible inventory policies (for all $t$). The problem is:

$\max_{\sigma}\mathbb{E}\Big[\sum_{t=0}^{T}\beta^{t}\pi_{t}\Big]$

But we want it recursively... 

* decisions are $q_{t}$
* what do we need to take into account when these decisions are made?
    - profit depends on current stock (x) and demand (d) $\rightarrow$ state variables 
    
$V(x_{t}, d_{t}) = \max_{q_{t} \geq 0}\Big\{\pi_{t} + \beta\mathbb{E}\Big[V(x_{t+1},d_{x+1})\Big|x_{t}, d_{t}, q_{t}\Big]\Big\}$

$V(x_{t}, d_{t}) = \max_{q_{t} \geq 0}\Big\{s_{t} p - k_{t} r - c\mathbb{1}\{q_{t} > 0\} + \beta\mathbb{E}\Big[V(x_{t+1},d_{x+1})\Big]\Big\}$ (because we assume that demand is idiosyncratic (random draw))

$s_{t} = \min\{x_{t}, d_{t}\}$

$k_{t} = \max\{x_{t} - d_{t}, 0\} + q_{t}$

Substitute the distribution of demand:

$V(x_{t}, d_{t}) = \max_{q_{t} \geq 0}\Big\{s_{t} p - k_{t} r - c\mathbb{1}\{q_{t} > 0\} + \beta \int V(k_{t}, d)\partial F(d) \Big\}$


$V(x_{t}, d_{t}) = \max_{q_{t} \geq 0}\Big\{s_{t} p - k_{t} r - c\mathbb{1}\{q_{t} > 0\} + \beta \sum_{d} V(k_{t}, d)pr(d)\Big\}$

#### Deterministic case

* let $d$ be fixed 

$V(x_{t}) = \max_{q_{t} \geq 0}\Big\{p \min\{x_{t}, d_{t}\} - r \max\{x_{t} - d_{t}, 0\} + q_{t} - c\mathbb{1}\{q_{t} > 0\} + \beta V\big(\max\{x_{t} - d, 0\} + q_{t}\big)\Big\}$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
class inventory_model:
    '''Small class to hold model fundamentals and its solution'''

    def __init__(self,label='noname',
                 max_inventory=10,  # upper bound on the state space
                 c = 3.2,            # fixed cost of order
                 p = 2.5,           # profit per unit of good
                 r = 0.5,           # storage cost per unit of good
                 β = 0.95,          # discount factor
                 demand = 4         # fixed demand
                 ):
        '''Create model with default parameters'''
        self.label=label # label for the model instance
        self.c, self.p, self.r, self.β = c, p, r, β
        self.demand = demand
        # created dependent attributes (it would be better to have them updated when underlying parameters change)
        self.n = max_inventory+1    # number of inventory levels
        self.upper = max_inventory  # upper boundary on inventory
        self.x = np.arange(self.n)  # all possible values of inventory (state space) (grid of all possible values)

    def __repr__(self):
        '''String representation of the model'''
        return 'Inventory model labeled "{}"\nParameters (c,p,r,β) = ({},{},{},{})\nDemand={}\nUpper bound on inventory {}' \
               .format (self.label,self.c,self.p,self.r,self.β,self.demand,self.upper)

    def sales(self,x,d):
        '''Sales in given period'''
        return np.minimum(x,d)

    def next_x(self,x,d,q):
        '''Inventory to be stored, becomes next period state'''
        return x - self.sales(x,d) + q

    def profit(self,x,d,q):
        '''Profit in given period'''
        return self.p * self.sales(x,d) - self.r * self.next_x(x,d,q) - self.c * (q>0)
```

```python
def bellman(m, v0):
    '''m = model object
    v0 = next period value function'''
    
    # grid of choices
    q = m.x[:, np.newaxis]
    
    # compute profit
    p = m.profit(m.x, m.demand, q)
    
    # indexes next period value
    i = np.minimum(m.next_x(m.x,m.demand,q),m.upper)
        # next period inventory (a matrix)
        # in the Bellman equation, we calculate value function as some qt (see it above)
    
    # bellman
    vm = p + m.β * v0[i] # still a matrix
    
    v1 = np.amax(vm, axis = 0) # maximum in every column
    q1 = np.argmax(vm, axis = 0) # max in every column, order volume 
    
    return v1, q1
```

### Backward induction algorithm

* start at $t = T$
* decrease the Bellman equations by a time period, unless $t = 1 \Rightarrow$ leading to a complete strategy

```python
def solver_backwards_induction(m,T=10,verbose=False):
    '''Backwards induction solver for the finite horizon case'''
    # solution is time dependent
    m.value  = np.zeros((m.n,T)) # initialise the value and policy as matrices 
    m.policy = np.zeros((m.n,T))
    # main DP loop (from T to 1)
    for t in range(T,0,-1):
        if verbose:
            print('Time period %d\n'%t)
        j = t-1 # index of value and policy functions for period t
        if t==T:
            # terminal period: ordering zero is optimal
            m.value[:,j] = m.profit(m.x,m.demand,np.zeros(m.n))
            m.policy[:,j] = np.zeros(m.n)
        else:
            # all other periods
            m.value[:,j], m.policy[:,j] = bellman(m,m.value[:,j+1]) # next period to Bellman
        if verbose:
            print(m.value,'\n')
    # return model with updated value and policy functions
    return m
```

```python
model = inventory_model(label='illustration')
model=solver_backwards_induction(model,T=5,verbose=True)
print('Optimal policy:\n',model.policy)

# x axis is the time period
# y axis is the state space (discrete)
```

```python
def plot_solution(model):
    plt.step(model.x,model.value)
    plt.legend([f'{i+1}' for i in range(model.value.shape[1])])
    plt.title('Value function')
    plt.show()
    plt.step(model.x,model.policy)
    plt.legend([f'{i+1}' for i in range(model.policy.shape[1])])
    plt.title('Policy function (optimal order sizes)')
    plt.show()

plot_solution(model)
```

# Coding the bus engine replacement model

* equations to be put into the code:
    - the expected value function (in expected value function space)
    - state space is given by $n$ grid points of mileage
    - choice space is given by two points
    - $\mathrm{EV}(x,d)$ is an $n x 2$ matrix 
        - dimension of $x$ and $d$
    - utility function 
    - transition matrix for mileage 
* use the Value Function Iteration algorithm 

--

* to simplify code:
    - $\mathrm{EV}(x,d=\mathrm{replace}) = \mathrm{EV}(0,d=\mathrm{keep})$ (EV of replacing is the same as keeping the new bus)
    - then use a single vector of $n$ elements (first component is the same as replacing)
    - then only need one transition probability matrix 
    - **logsum** functions for large numbers, use a stabilising technique of adding $M$ (take constant out of values in exponent) 
        - good choice of $M$ is max. of numbers be have to put into **logsum**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
plt.rcParams['figure.figsize'] = [12, 8]

class zurcher():
    '''Harold Zurcher bus engine replacement model class, VFI version'''
    
    
    def __init__(self, 
                 n = 175, # state points 
                 RC = 11.7257, # replacement cost
                 c = 2.45569, # parameter of maintenance cost (θ_1)
                 p = [0.0937, 0.4475, 0.4459, 0.0127], # probabilities of transition (θ_2)
                 β = 0.9999): 
        '''Init for the Zurcher model object'''
    
        assert sum(p) <= 1.0, 'Transition probability parameters must sum up to <1'
    
        self.RC, self.c, self.p, self.β, self.n = RC, c, p, β, n
        
        self.grid = np.arange(n) # list of integers that can act as a grid 
        
        self.trpr = self.__transition_probs()
       
    
    @property
    def n(self):
        '''Attribute getter for n'''
        return self.__n
    
    
    @n.setter
    def n(self, value): # called everytime we need n (then define grid and θ_2 in here)
        '''Attribute n setter'''
        
        self.__n = value
        
        self.grid = np.arange(self.__n) # list of integers that can act as a grid 
        
        self.trpr = self.__transition_probs()
        
        
    def __repr__(self):
        '''String representation of the Zurcher model'''
        return 'Rust model at id={}'.format(id(self))
        
        
    def __transition_probs(self):
        '''Computing the transition probability matrix'''
        trpr = np.zeros((self.__n, self.__n)) # we need to check if rows add up to one
        
        probs = self.p + [1 - sum(self.p)] # ensure sum up to 1

        for i, p in enumerate(probs):
            trpr += np.diag([p] * (self.__n - i), k = i)
            
        trpr[:,-1] = 1. - np.sum(trpr[:,:-1], axis = 1)
        
        return trpr
    
    
    def bellman(self, ev0): # EV in previous period
        x = self.grid # points in the next period state 
        
        mcost = -0.001 * x * self.c  # maintenance cost
        vx0 = mcost + self.β * ev0 # d = 0 (if not replacing then we keep going -- problem does not start again)
        # vector of v(x,0)
        vx1 = -self.RC + mcost[0] + self.β * ev0[0] # d = 1 (adding the trick from before)
        
        M = np.maximum(vx0, vx1)
        logsum = M + np.log(np.exp(vx0 - M) + np.exp(vx1 - M))
        
        ev1 = self.trpr @ logsum[:,np.newaxis] # this is the bellman
        ev1 = ev1.ravel()
        
        # choice probabilities from the logit formula (recompute v for current period choice probabilities)
        vx0 = mcost + self.β * ev1
        vx1 = -self.RC + mcost[0] + self.β * ev1[0]
        pr1 = 1 / (np.exp(vx0 - vx1) + 1) # divide denominator and numerator by exp(vx1)
        
        return ev1, pr1 # T(ev0), and policy choice (choice probability of choosing an action)
    
    
    def solve_vfi(self, tol = 1e-6, maxiter = 100, callback = None):
        '''Solves the Rust model using value function iterations'''
    
        ev0 = np.zeros(self.n)
        for i in range(maxiter):
            ev1, pr1 = self.bellman(ev0)
            err = np.amax(np.abs(ev0 - ev1)) # allows x to be an array
            if callback != None: callback(iter = i, err = err, ev1 = ev1, ev0 = ev0, pr1 = pr1, model = self)
            if err < tol:
                break
            ev0 = ev1
        else:
            raise RuntimeError('Failed to converge in %d iterations' % maxiter)
        return ev1, pr1
    
    
    def solve_show(self,maxiter=1000,tol=1e-6,**kvargs):
        '''Illustrate solution'''
        fig1, (ax1,ax2) = plt.subplots(1,2,figsize=(14,8))
        ax1.grid(b=True, which='both', color='0.65', linestyle='-')
        ax2.grid(b=True, which='both', color='0.65', linestyle='-')
        ax1.set_xlabel('Mileage grid')
        ax2.set_xlabel('Mileage grid')
        ax1.set_title('Value function')
        ax2.set_title('Probability of replacing the engine')
        def callback(**argvars):
            mod, ev, pk = argvars['model'],argvars['ev1'],argvars['pr1']
            ax1.plot(mod.grid,ev,color='k',alpha=0.25)
            ax2.plot(mod.grid,pk,color='k',alpha=0.25)
        ev,pk = self.solve_vfi(maxiter=maxiter,tol=tol,callback=callback,**kvargs)
        # add solutions
        ax1.plot(self.grid,ev,color='r',linewidth=2.5)
        ax2.plot(self.grid,pk,color='r',linewidth=2.5)
        plt.show()
        return ev,pk
        
    

```

# Cake eating in a discrete world 

* initial size $W_{0}$
* how much cake to eat in each period?
* $W_{t+1} = W_{t} - c_{t}$

Model specification:
* choice set/ choices: $c_{t}$
* state space: $W_{t}$
* preferences: $u(\cdot)$
* beliefs about how the state evolves: $W_{t+1}$ from above
* policy function: $c^{*}(W_{t})$

$u(c_{t}) = \log(c_{t})$

* cake will never be eaten completely, since $c_{t} = 0$ gives $- \infty$ utility
* we want to maximise the discounted expected lifetime utility 

The Bellman equation:
$V(W_{t}) = \max_{0 \leq c_{t} \leq W_{t}}\Big\{u(c_{t}) + \beta V(W_{t+1})\Big\}$

where $W_{t+1} = W_{t} - c_{t}$.

--

Start with a guess of $V(W) = A + B\log W$:

$A + B\log W = \max_{c}\Big\{\log c + \beta(A + B\log(W - c)\Big\}$

Take the FOC wrt. $c$:

$\frac{1}{b} + \beta B \frac{1}{W - c}(-1) = 0$

$c\beta B = w - c$

$c = \frac{w}{1 + \beta B}$

$W - c = \frac{\beta B}{1 + \beta B}W$

Then it follows that:

$A + B\log W = \log W - \log (1 + \beta B) + \beta A + \beta B \log(\frac{\beta B W}{1 + \beta B})$

$A + B\log W = \log W - \log (1 + \beta B) + \beta A + \beta B \log \beta B - \beta B \log(1 + \beta B) + \beta B \log W$

$A + B\log W = \log W(1 + \beta B) + \Psi$


Therefore, we have:

$B = 1 + \beta B \Rightarrow B = \frac{1}{1 - \beta}$

$A$ recovered in similar way, but more complex... We can then easily solve for $V(W)$ and $c^{*} = (1 - \beta)W$.

--

The Bellman operator $T(V)(W)$ and then we look for the fixed point (using VFI). 



The cake size is continuous. This is difficult to implement with the Bellman operator. **Solution**: we discretize $W$:

$\mathbf{W} \in \{0,...,\bar{W}\}$

* we compute the value and policy function sequentially point-by-point (interpolation)

Can we avoid interpolation here?

* note, choice $c_{t}$ determines the next period $W_{t+1}$... what if we swap them in the Bellman equation so that the next period cake size is the decision variable (solving 'on the grid'). However, we use $\mathbf{W}$ twice: in the state space and the decision space... problem? We will come back to it at the end of the class...

```python
import numpy as np

class cake_ongrid():
    
    def __init__(self, β = 0.9, Wbar = 10, ngrid = 50):
        '''Initializer'''
        
        self.β = β
        self.Wbar = Wbar
        self.ngrid = ngrid
        self.ϵ = np.finfo(float).eps # gives the smallest positive float number (smallest difference between two floats)
        self.grid = np.linspace(self.ϵ, Wbar, ngrid)
        
    
    def bellman(self, V0):
        c = self.grid - self.grid[:,np.newaxis] # consumption today (state today - choice)
        # current states in columns and choices in rows 
        
        c[c==0] = self.ϵ # add small quantity to avoid log(0)
        
        mask = c>0 # mask off (do not select) infeasible choices (i.e. negative consumption)
        
        matV1 = np.full((self.ngrid,self.ngrid),-np.inf) # init V with -inf
        # matrix of values that we want to output 
        matV0 = np.repeat(V0.reshape(self.ngrid,1),self.ngrid,1) #current value function repeated in columns
        # reshape V0 into a column vector, and want to repeat it n times (so that it is a matrix)
        # 1 implies the axis; so we get horizontally stacked column vectors that are all the same 
        
        matV1[mask] = np.log(c[mask]) + self.β * matV0[mask] # maximand of the Bellman equation
        
        V1 = np.amax(matV1, axis = 0, keepdims = False) # maximum in every column
        c1 = self.grid - self.grid[np.argmax(matV1, axis = 0)] # consumption (index of maximum in every column)
        return V1, c1
        
        
    def solve(self, maxiter = 100, tol = 1e-4, callback = None):
        V0 = np.log(self.grid)
        for iter in range(maxiter):
            V1, c1 = self.bellman(V0)
            if callback: callback(iter, self.grid, V1, c1)
            if np.all(abs(V1 - V0) < tol):
                break
            V0 = V1
        else:
            print('No convergence: maximum number of iterations achieved!')
        return V1, c1
```

```python
model = cake_ongrid(β = 0.92, Wbar = 10, ngrid = 500)
V, c = model.solve()
print(c)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['axes.autolimit_mode'] = 'round_numbers'
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0
plt.rcParams['patch.force_edgecolor'] = True
from cycler import cycler
plt.rcParams['axes.prop_cycle'] = cycler(color='bgrcmyk')

fig1, ax1 = plt.subplots(figsize=(12,8))
plt.grid(b=True, which='both', color='0.65', linestyle='-')
ax1.set_title('Value function convergence with VFI')
ax1.set_xlabel('Cake size, W')
ax1.set_ylabel('Value function')

def callback(iter,grid,v,c):
    '''Callback function for DP solver'''
    if iter<5 or iter%10==0:
        ax1.plot(grid[1:],v[1:],label='iter={:3d}'.format(iter),linewidth=1.5)

V,c = model.solve(callback=callback, maxiter = 1000)

plt.legend(loc=4)
# plt.savefig('cake1value.eps', format='eps', dpi=300)
plt.show()
```

## Meaurement errors 

* compared to the analytical solution
* if no analytical solution, can derive theoretical property of the solution, and then can see if the value function/ solution satisfies those properties

```python
fun = lambda w: np.log(w)/(1 - model.β) + np.log(1 - model.β)/(1 - model.β) + model.β*np.log(model.β)/((1 - model.β)**2)
grid = model.grid
plt.plot(grid[1:],V[1:],linewidth=1.5)
plt.plot(grid[1:],fun(grid[1:]),linewidth=1.5)
```

```python
apol = lambda w: (1 - model.β) * w
grid = model.grid
plt.plot(grid[1:],c[1:],linewidth=1.5)
plt.plot(grid[1:],apol(grid[1:]),linewidth=1.5)
```
