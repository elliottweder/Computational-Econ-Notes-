# Linear Regression with Numpy and Pandas 

$\hat{\beta} = (X'X)^{-1}X'y$
 
* we also want to estimate the accuracy of the estimates - look at the variance-covariance matrix too:
    - $\hat{\sum} = s^{2}(X'X)^{-1}$
    
* where $s = \sqrt{\frac{1}{n - k}e'e}$
    - n is observations and k is the number of regressors 

```python
import numpy as np

def ols(X, y, addConstant=True, verbose=True):
    
    if X.ndim==1:
        X = X[:, np.newaxis] # converting 1D array into a column vector 
        
    if y.ndim==1:
        y = y[:, np.newaxis]
        
    
    if addConstant:
        # add a column of ones to X
        X = np.hstack((np.ones((X.shape[0],1)), X)) # X.shape[0] is the number of columns 
    
    beta = np.linalg.inv(X.T @ X) @ X.T @ y
    
    e = y - X @ beta 
    
    n, k = X.shape
    
    s2 = e.T @ e / (n - k)
    
    s = np.sqrt(s2)[0,0] # because np.sqrt is still an array, so just take the first part of it (which is actually the same)
    
    Sigma = np.linalg.inv(X.T @ X) * s2
    
    beta_std = np.sqrt(np.diag(Sigma))
    
    if verbose:
        print('Number of observations: {}'.format(n))
        print('Number of parameters: {}'.format(k))
        # print(s, type(s), s.ndim, s.shape)
        print('MSE = {:.5f}'.format(s))
        
        for b,std in zip(beta, beta_std):
            b = b[0]
            print('{:10.5f} ({:10.5f})'.format(b, std))
    
    return beta, Sigma
```

```python
X = np.array([[5, 3],
              [2, 3],
              [3, 1],
              [2, 8],
              [4.5, 2.5],
              [2.5, 1.5],
              [4.3, 4.2],
              [0.5, 3.5],
              [1, 5],
              [3, 8]]) # 2D array organised in columns 


truebeta = np.array([1.234,-0.345])[:,np.newaxis]  # column vector (newaxis tells numpy to be arranged with an additional (
# the second axis))
# want a column vector to multiply here 


y = X @ truebeta + 2.5 + np.random.normal(size=(X.shape[0],1),scale=0.2) # this is the real equation

# print(X, truebeta, y, sep='\n\n')

beta,S=ols(X,y)
# print(beta, S, sep='\n\n')
beta,S=ols(X,y,addConstant=False)
```

```python
def ols2(X,y,addConstant=True,verbose=True):
    '''Return the OLS estimates and their variance-covariance matrix for the given data X,y
    When addConstant is True, constant is added to X
    When verbose is True, a report is printed
    '''
    y = y.squeeze()  # we are better off if y is one-dimensional
    if addConstant and X.ndim==1:
        X = np.hstack((np.ones(X.shape[0])[:,np.newaxis],X[:,np.newaxis]))
        k = 2
    elif addConstant and X.ndim>1:
        X = np.hstack((np.ones(X.shape[0])[:,np.newaxis],X))
        k = X.shape[1]+1
    elif X.ndim==1:
        X = X[:,np.newaxis]
    xxinv = np.linalg.inv(X.T@X)  # inv(X'X): pre calculation is good for efficiency 
    beta = xxinv @ X.T@y  # OLS estimates
    e = y - X@beta  # residuals
    n,k = X.shape  # number of observations and parameters
    s2 = e.T@e / (n-k) # is just a scalar, because y is uni-dimensional 
    Sigma = s2*xxinv
    if verbose:
        # report the estimates
        print('Number of observations: {:d}\nNumber of parameters: {:d}'.format(n,k))
        print('Parameter estimates (std in brackets)')
        for b,s in zip(beta,np.sqrt(np.diag(Sigma))):
            print('{:10.5f} ({:10.5f})'.format(b,s))
        print('MSE = {:1.5f}\n'.format(np.sqrt(s2)))
    return beta,Sigma
```

# Linear Programming and Optimal Transport Models 

* finding max/ min of linear functions subject to a set of linear inequality constraints 

--

Optimal transport problem: minimum cost of transporting a set of goods from *m* origins to *n* destinations, where the cost of transporting from origin *i* to destination *j* is given by $c_{ij}$ (in a matrix C). In economics:

* matching and trade:
    - matching distribution of workers and firms 
* demand estimation and pricing 
    - discrete choice 
* quantile methods in econometrics 

* columns of matrix are destinations, rows and the origins 

min $\sum_{i=1}^{m}\sum_{j=1}^{n}c_{ij}x_{ij}$ subject to:

$\sum_{i=1}^{m}x_{ij} = b_{j}$ (all goods that arrive at *j* have to exhaust demand there)
    
$\sum_{j=1}^{n}x_{ij} = a_{i}$ (all goods that leave *i* have to exhaust amount of goods there)

Then we translate this into:

max($c \cdot x$) subject to $Ax \leq b$

* Row of A multiplied by column vector x gives us a line - and we want it to be less than the line b (element of b)
    - line a line for a budget constraint
    - we form a convex polygon using this method 

Convex polyhedron in 3D (with 3 variables)
* we move the level curve of the objective function to find the vertex of the 3D shape (this is the process of maximising)

## Example: Optimal production portfolio

* production technology is restricted to have:
   
$ y - x \leq 4$

$ 2x - y \leq 8$

(like complementary goods - the difference between the two goods cannot be too large in each direction)

* the resource constraint is given by $x + 2y \leq 14$
* profits are given by $\pi(x, y) = y + 2x$

All equations are linear - convert to linear programming problem:

max $(c^{T}x)$ subject to $Ax \leq b$ with: 

c = (2 1)

A = ((-1 1), (2 -1), (1 2), (-1 0), (0, -1))

b = (4, 8, 14, 0, 0)'

* two bottom rows of A incorporate the non-negativity constraints 

```python
import numpy as np
from scipy.optimize import linprog

c = np.array([-2,-1])
A = np.array([[-1,1],[2,-1],[1,2],[-1,0],[0,-1]])
b = np.array([4,8,14,0,0])

# put (-) sign infront because we are maximising, but the solver is minimising 

def outf(arg):
    print('iteration %d, current solution %s' % (arg.nit, arg.x))
    
linprog(c = c, A_ub = A, b_ub = b, method = 'simplex', callback=outf)
```

# Measuring Illegal Trade with Linear Programming 

* measuring black market activity
* introduction of regulation on Beijing licence plates (licence plates assigned by lottery to potential owners)
* allocation by random lottery should have no effect on car sales; but there is a sizeable shift in the distribution of cars
* the number of cars should change
    - want to measure a lower bound on the volume of illegal trade of licence plates 
    

* licence plates were allocated via a lottery (quota)
* you must have a Beijing licence plate to drive in Beijing
* a Beijing household needs a licence plate before it can register for a new car
* licence plates are non-transferable 

--

* shift in distribution is from cheaper to more expensive car models 
    - if lottery random: then same sorts of people buying cars in the past should be buying them now (just fewer people)
* it is rich people getting plates 
* no similar shifts in sales in comparable cities without rationing policy in the same time period
* no supply side responses to the rationing policy

## Modelling framework 

* let $P_{0}$ be the dist. pre-lottery (and $P_{1}$)
* under assumptions:
    - pricing policy did not change 
    - demand structure did not change 
    - lottery is uniform
The distribution of new car sales should not change.

Do we see a change in the distribution? 

If we have less bins, we can see that the second distribution (2011) has a higher density of more expensive cars (around 200k)

Set up a transportation problem: minimum carry from one distribution to the other to explain the change we see above

* stack all $x_{ij}$ into a single vector (cost matrix c too)

--

* origin is the original distribution, and destination is the new dist.

* how do we fix the distributions?

```python
N = 20 # number of bins to represent the distribution 

dt['bin'] = pd.qcut(dt.MSRP, q=N, labels=False) # want to get an integer/ indices that tells us the bin (rather than the 
# parentheses); so add labels 

# want fraction in each bin
gr10 = dt[dt['year']==2010].groupby('bin')
gr11 = dt[dt['year']==2011].groupby('bin')

print(gr10.MSRP.count())
print(gr11.MSRP.count())
```

```python
# need to convert into fractions, rather than just the count (total number of car purchases fell)
```

```python
d10 = gr10.MSRP.count() / dt[dt['year']==2010]['year'].count()
d11 = gr11.MSRP.count() / dt[dt['year']==2011]['year'].count()

print(d10, d11)
```

```python
# set up the transportation problem 

import numpy as np 

costs = np.ones((N, N)) - np.eye(N) # it is free to leave the dist. as it is, and costly to move it elsewhere 
# print(costs); costly to move from any bin to another 
# by minimising problem, we are finding lower bound (least amount of prob. mass transfer that is needed to explain 
# the change in dist.)

origins = np.array(d10)

destinations = np.array(d11)
```

```python
# convert to a linear programming problem 

c = costs.reshape(N * N) # it is 1D

A1 = np.kron(np.ones((1,N)), np.eye(N))
# imagine we are stacking $x_{ij}$ by each columns - so N x N is now a N*N x 1 matrix 
# then A should have 1s N apart - origin
# next row should have 1s N apart but over one column compared to above 

A2 = np.kron(np.eye(N), np.ones((1,N)))

A = np.vstack((A1, A2))

#plt.spy(A1) (shows non-zero elements of a matrix)
#plt.spy(A2)
#plt.spy(A)

b = np.concatenate((origins, destinations))
```

```python
# linear programming problem solver 

from scipy.optimize import linprog

res = linprog(c = c, A_eq = A[:-1], b_eq = b[:-1], method='simplex')
```

A does not appear to be full rank - once N - 1 is satisfied, then the last is 

fun - value of the function 

```python
x = res.x.reshape((N, N))

print(x)

plt.spy(x)

estimate = 1 - np.diag(x).sum()

print('The lower bound on the share of illegal trade is %1.5f%%' % (estimate*100))
```

The off-diagonal elements are where the transitions happen (from low brings into higher bins). No probability mass transfer in diagonals.
